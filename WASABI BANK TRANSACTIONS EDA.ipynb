```python
import pandas as pd
import numpy as np
```

```python
# unioned transaction data is loaded
atm_transactions = pd.read_csv("fulltransactiondata.csv")
```

```python
# matplotlib and seaborn are imported
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Optional: Set plot style
sns.set(style="whitegrid")
```

```python
#A WHEN time-based Analysis
#Here the Peak hours is being analyzed
hourly_usage = atm_transactions['hour'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
sns.barplot(x=hourly_usage.index, y=hourly_usage.values, color='skyblue')
plt.title('Transactions by Hour of the Day')
plt.xlabel('Hour (0â€“23)')
plt.ylabel('Number of Transactions')
plt.xticks(range(24))
plt.tight_layout()
plt.show()
```

```python
#Weekday VS Weekends
weekend_usage = atm_transactions['Day Name'].value_counts().reindex(
    ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
)

plt.figure(figsize=(10, 6))
sns.barplot(x=weekend_usage.index, y=weekend_usage.values, palette="viridis")
plt.title('Transactions by Day of the Week')
plt.xlabel('Day')
plt.ylabel('Number of Transactions')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Holiday VS Non-Holiday
plt.figure(figsize=(6, 4))
sns.countplot(data=atm_transactions, x='IsHoliday', palette='Set2')
plt.title('Holiday vs Non-Holiday Transactions')
plt.xlabel('Is Holiday')
plt.ylabel('Number of Transactions')
plt.tight_layout()
plt.show()
```

```python
#WHERE - LOCATION ANALYSIS
#Top and Bottom ATMs
top_atms = atm_transactions['Location Name'].value_counts().nlargest(10)
bottom_atms = atm_transactions['Location Name'].value_counts().nsmallest(10)

# Top
plt.figure(figsize=(10, 6))
sns.barplot(y=top_atms.index, x=top_atms.values, color='green')
plt.title('Top 10 Performing ATM Locations')
plt.xlabel('Transactions')
plt.ylabel('ATM Location')
plt.tight_layout()
plt.show()

# Bottom
plt.figure(figsize=(10, 6))
sns.barplot(y=bottom_atms.index, x=bottom_atms.values, color='red')
plt.title('Bottom 10 ATM Locations')
plt.xlabel('Transactions')
plt.ylabel('ATM Location')
plt.tight_layout()
plt.show()
```

```python
#Urban Vs Rural (State-Level Comparison)
state_usage = atm_transactions['State'].value_counts().nlargest(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=state_usage.values, y=state_usage.index)
plt.title('Top 10 States by ATM Usage')
plt.xlabel('Transactions')
plt.ylabel('State')
plt.tight_layout()
plt.show()
```

```python
# WHO Demographic Usage
# Gender Usage
plt.figure(figsize=(6, 4))
sns.countplot(data=atm_transactions, x='Gender', palette='pastel')
plt.title('Transactions by Gender')
plt.xlabel('Gender')
plt.ylabel('Transactions')
plt.tight_layout()
plt.show()
```

```python
#Account Type
plt.figure(figsize=(6, 4))
sns.countplot(data=atm_transactions, x='AccountType', palette='muted')
plt.title('Transactions by Account Type')
plt.xlabel('Account Type')
plt.ylabel('Transactions')
plt.tight_layout()
plt.show()
```

```python
#Occupation Analysis Top(10)
top_occupations = atm_transactions['Occupation'].value_counts().nlargest(10)

plt.figure(figsize=(10, 6))
sns.barplot(y=top_occupations.index, x=top_occupations.values, color='purple')
plt.title('Top 10 Occupations by ATM Usage')
plt.xlabel('Transactions')
plt.ylabel('Occupation')
plt.tight_layout()
plt.show()
```
